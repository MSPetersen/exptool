```python
import yt
import psp_io
import numpy as np
```

```python
O = psp_io.Input('/Users/mpetersen/Research/NBody/Disk064a/OUT.run064a.01000',comp='star',nout=100000)
data = {'particle_position_x': O.xpos, 'particle_position_y': O.ypos,'particle_position_z': O.zpos,'particle_mass': O.mass}
bbox = 1.1*np.array([[min(O.xpos), max(O.xpos)], [min(O.ypos), max(O.ypos)], [min(O.zpos), max(O.zpos)]])
ds = yt.load_particles(data,n_ref=256,bbox=bbox)

```

```python
ad = ds.all_data()
cic_density = ad["deposit", "all_cic"]
nn_density = ad["deposit", "all_density"]
nn_deposited_mass = ad["deposit", "all_mass"]
particle_count_per_cell = ad["deposit", "all_count"]

```

```python
#p = yt.SlicePlot(ds, "z", "all_density")
#p = yt.ProjectionPlot(ds, "z", "all_cic")
sc = yt.create_scene(ds,field=('all_density'), lens_type='perspective')


```

```python
sc.camera.set_width(ds.quan(20, 'kpc'))

source = sc.sources['source_00']

tf = yt.ColorTransferFunction((-28, -24))
tf.add_layers(4, w=0.01)

source.set_transfer_function(tf)
sc.show()
```
